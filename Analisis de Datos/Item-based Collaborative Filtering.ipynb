```python
import pymongo
client = pymongo.MongoClient('mongodb://anal:anal18@18.220.137.254/anal') 
db = client['anal']
usersdb = db['users']
animesdb = db['animes']
tagsdb = db['tags']
```

```python
import pandas as pd
users= usersdb.find()
usersList = list(users)
n_users= len(usersList)
x=0
for userInd in range(n_users):
    user=usersList[userInd-x]
    animeWatched = user['animes']
    i=0
    for j in range(len(animeWatched)):
        anime=animeWatched[j-i]        
        if anime['ratingUser']==0:
            #print("gonna remove: "+ str(anime['animeId']))
            animeWatched.remove(anime)
            i=i+1
    if len(animeWatched)<50:
        usersList.remove(user)
        x=x+1        
    
usersdf = pd.DataFrame(usersList)

print(usersdf)
```

```python
animes= animesdb.find()
animesdf = pd.DataFrame(list(animes))
animesdf = animesdf.drop('tags',1).drop('rating',1).rename(index=str, columns={"_id": "animeId"})
n_animes= len(animesdf)

print(animesdf)
```

```python
#Se crea la estructura : 'animeId' 'ratingUser' 'userId'. Es decir el rating que cada usuario le da a cada anime que ha visto
animesWdf=  pd.DataFrame(list(usersdf.iloc[0]['animes']))
animesWdf['userId']=usersdf.iloc[0]['_id']

cantUsers=len(usersdf)-1 #-1 del primer usuario analizado arriba
for i in range(cantUsers):
    xd=  pd.DataFrame(list(usersdf.iloc[i+1]['animes']))
    xd['userId']=usersdf.iloc[i+1]['_id']
    animesWdf = pd.concat([animesWdf, xd],ignore_index=True)
    
print(animesWdf)
```

```python
#Mergeamos las dos tablas
ratings = pd.merge(animesdf, animesWdf)
```

```python
print(ratings)
```

```python
#Creamos la tabla pivote donde podemos ver el rating que cada usuario le ha dado a cada anime
animeRatings = ratings.pivot_table(index=['userId'],columns=['title'],values='ratingUser')
animeRatings.head()
```

```python
import numpy as np
# agregamos por titulo y devolvemos el 
# numero de veces que se puntuo, y la media de puntuacion
animeStats = ratings.groupby('title').agg({'ratingUser': [np.size, np.mean]})
 
# nos quedamos con todas las que tengan mas de 100 puntuaciones 
# de distintos usuarios
popularAnimes = animeStats['ratingUser']['size']> 100
 
# ordenamos por la puntuación asignada
animeStats[popularAnimes].sort_values([('ratingUser', 'mean')], ascending=False)[:15]
```

```python
# correlamos todas las columnas con todas las demás
# usamos el metodo pearson
# descartamos todas aquellas que no tengan al menos 100 valoraciones de usuarios
corrMatrix = animeRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = animeRatings.iloc[1].dropna().sort_values(ascending=False)
print(myRatings)
```

```python
posiblesSimilares = pd.Series()
 
# Recorremos los animes valorados por el usuario
for i in range(0, len(myRatings.index)):
    # Obtenemos animes similares a este que el usuario ha puntuado
    sims = corrMatrix[myRatings.index[i]].dropna()

    # Multiplicamos el score de correlacion por la puntuación asignada por el usuario
    sims = sims.map(lambda x: x * myRatings[i])

    # Añadimos el anime y la nueva puntuacion a nuestra lista de candidatos
    posiblesSimilares = posiblesSimilares.append(sims)

# Agrupamos los resultados, ya que si un anime es parecida a dos de los que ha visto el usuario, aparecerá dos veces
# podríamos agrupar haciendo la media de puntuación, o como en este caso sumandolo, ya que si aparece dos veces será
# porque es un anime muy recomendable... y así saldrá primero. Va en cuestión de gustos.
posiblesSimilares = posiblesSimilares.groupby(posiblesSimilares.index).sum()

# Finalmente filtramos todos aquellos animes que el usuario ya habia valorado puesto que no tiene sentido que se 
# las recomendemos si ya las ha visto
filtered = posiblesSimilares.drop(myRatings.index,errors='ignore')
filtered = filtered.sort_values(ascending=False)
print(filtered)
#print(filtered.head(10))
```
