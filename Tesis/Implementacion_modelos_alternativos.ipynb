## Splitting the data

```
from sklearn.model_selection import train_test_split

y= df2.Lima
X= df2.drop('Lima',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

## Classifiers

```
from sklearn.metrics import confusion_matrix
def getAccRecall(y_test,y_pred):
  tn, fp, fn, tp = confusion_matrix(y_test, y_pred, labels=[0,1]).ravel()
  accuracy = (tn + tp) / (tn+tp+fp+fn)
  recall = tp/(tp+fn)
  accuracy = round(accuracy*100,2)
  recall = round(recall*100,2)
  return (accuracy,recall)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score

model_LR = LogisticRegression(random_state=0).fit(X_train, y_train)
model_KNN = KNeighborsClassifier(n_neighbors=50).fit(X_train,y_train)
model_SVM_rbf = SVC(kernel='rbf').fit(X_train,y_train)
model_SVM_poly = SVC(kernel='poly').fit(X_train,y_train)
model_DTC = DecisionTreeClassifier(max_depth=5).fit(X_train,y_train)
model_RFC = RandomForestClassifier(max_depth=5).fit(X_train,y_train)
model_ABC = AdaBoostClassifier().fit(X_train, y_train)
model_GNB = GaussianNB().fit(X_train, y_train)

y_pred_LR = model_LR.predict(X_test)
y_pred_KNN = model_KNN.predict(X_test)
y_pred_SVM = model_SVM_rbf.predict(X_test)
y_pred_DTC = model_DTC.predict(X_test)
y_pred_RFC = model_RFC.predict(X_test)
y_pred_ABC = model_ABC.predict(X_test)
y_pred_GNB = model_GNB.predict(X_test)

acc_rec_LR = getAccRecall(y_test,y_pred_LR)
acc_rec_KNN = getAccRecall(y_test,y_pred_KNN)
acc_rec_SVM = getAccRecall(y_test,y_pred_SVM)
acc_rec_DTC = getAccRecall(y_test,y_pred_DTC)
acc_rec_RFC = getAccRecall(y_test,y_pred_RFC)
acc_rec_ABC = getAccRecall(y_test,y_pred_ABC)
acc_rec_GNB = getAccRecall(y_test,y_pred_GNB)
```

```
print(acc_rec_LR)
print(acc_rec_KNN)
print(acc_rec_SVM)
print(acc_rec_DTC)
print(acc_rec_RFC)
print(acc_rec_ABC)
print(acc_rec_GNB)
```

### SVM Models

#### SVM with RBF kernel

```
%%time
from sklearn.model_selection import cross_val_score

import numpy as np

bestScore=0.0
bestC=10000
bestG=10000
results=[]
for C in [0.5,1,5,10]:
  for G in np.linspace(0.1,5,50):
    model_SVM_rbf = SVC(kernel='rbf',C=C,gamma=G)
    scores = cross_val_score(model_SVM_rbf,X,y,cv=10)
    res = scores.mean()
    results.append([res,C,G])
    if res> bestScore:
      bestScore= res
      bestC=C
      bestG=G
```

```

print("Best Score: ",bestScore)
print("Best C: ",bestC)
print("Best G: ",bestG)
#BestC: 0.5 with BestG=0.2
```

```
model_SVM_rbf = SVC(kernel='rbf',C=0.5,gamma=0.2)
scores = cross_val_score(model_SVM_rbf,X,y,cv=10)
print(scores.mean())
```

#### SVM with polynomial kernel

```
%%time

bestScore=0.0
bestC=10000
bestD=10000
results=[]
for C in [0.5,1,5,10]:
  for D in [1,2,3,4]:
    model_SVM_poly = SVC(kernel='poly',C=C,degree=D)
    scores = cross_val_score(model_SVM_poly,X,y,cv=10)
    res = scores.mean()
    results.append([res,C,D])
    if res> bestScore:
      bestScore= res
      bestC=C
      bestD=D
```

```
print("Best Score: ",bestScore)
print("Best C: ",bestC)
print("Best D: ",bestD)
#BestC: 10 with BestD: 2
```

```
%%time
model_SVM_poly = SVC(kernel='poly',C=10,degree=2)
scores = cross_val_score(model_SVM_poly,X,y,cv=10)
print(scores.mean())
```

### MLP Model

```
%%time
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score

bestScore=0
best_lr= 0
best_layer_size= 0
resultsNN=[]
for lr in range(1,10):
  for hlayer_size in range(10,110,10):
      model_MLP = MLPClassifier(solver='adam', activation='tanh', hidden_layer_sizes=(hlayer_size), learning_rate_init=lr/10, max_iter=10000)
      scores = cross_val_score(model_MLP,X,y,cv=10)
      res= scores.mean()
      print("Con learning rat, hlayer_size: ", lr/10 ,hlayer_size, ". Se obtuvo de acc: ", res)
      resultsNN.append([res,lr/10,hlayer_size])
      if res>bestScore:
        bestScore=res
        best_lr=lr/10
        best_layer_size=hlayer_size
```

```
print("Best Score: ",bestScore)
print("Best learning rate: ",best_lr)
print("Best layer size: ",best_layer_size)
#BestC: 10 with BestD: 2
```
