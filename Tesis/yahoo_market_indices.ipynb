### Eliminamos los registros con algun indice nulo

```
df = df.dropna(axis=0)
df.sort_index(inplace=True, ascending=False)
```

```
df
```

### Transformamos los indices a valores discretos donde: 
####  0=disminucion del indice respecto a la fecha anterior
####  1=aumento del indice respecto a la fecha anterior
####  2=no hubo variacion

```
#0 ha disminuido
#1 ha aumentado
#2 se ha mantenido igual
rowAnt= df.loc[df.index[0]]
indexAnt=df.index[0]
for index, row in df.iterrows():
  if index==df.index[0]:
    continue
  for column in df.columns:
    if row[column]>rowAnt[column]:
      df.at[indexAnt,column] = 0
    elif row[column]<rowAnt[column]:
      df.at[indexAnt,column] = 1
    elif row[column]==rowAnt[column]:
      df.at[indexAnt,column] = 2
  indexAnt=index
  rowAnt=row
```
