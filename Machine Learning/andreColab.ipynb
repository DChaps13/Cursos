# Sistema de Recomendacion (Collaborative Filtering)

```
from keras.models import Model
from keras.callbacks import ModelCheckpoint
from keras.layers import dot, concatenate, Embedding, Input, Flatten, Dropout, Dense
import numpy as np
from sklearn.metrics import mean_absolute_error
```

```
EMBEDDING_SIZE = 100
VERBOSE = 1
```

```
animeUserList = pd.merge(animeUserList, users[['username', 'user_id']], how='left', on='username')
```

```
from sklearn.model_selection import train_test_split

ratings_train, ratings_test = train_test_split(animeUserList, test_size=0.2, random_state=0)
```

```
user_id_train = ratings_train['user_id']
item_id_train = ratings_train['anime_id']
rating_train = ratings_train['my_score']

user_id_test = ratings_test['user_id']
item_id_test = ratings_test['anime_id']
rating_test = ratings_test['my_score']

max_user_id = users['user_id'].max()
max_item_id = animeUserList['anime_id'].max()
```

```
def create_model(max_user_id, max_item_id):
  user_id_input = Input(shape=[1], name='user')
  item_id_input = Input(shape=[1], name='item')

  user_embedding = Embedding(output_dim=EMBEDDING_SIZE, input_dim=max_user_id + 1,
                             input_length=1, name='user_embedding')(user_id_input)
  item_embedding = Embedding(output_dim=EMBEDDING_SIZE, input_dim=max_item_id + 1,
                             input_length=1, name='item_embedding')(item_id_input)

  user_vecs = Flatten()(user_embedding)
  item_vecs = Flatten()(item_embedding)

  y = dot([user_vecs, item_vecs], axes=1)

  model = Model(inputs=[user_id_input, item_id_input], outputs=[y])
  model.compile(optimizer='adam', loss='mae')

  return model
```

```
rating_train.shape
```

```
batch_size=5
epoches=5
validation_split=0.1

model = create_model(max_user_id,max_item_id)

history = model.fit([user_id_train, item_id_train], rating_train, batch_size=batch_size, epochs=epoches, validation_split=validation_split, shuffle=True, verbose=VERBOSE)
```

```
def predict(model, user_ids, item_ids):
        predicted = model.predict([user_ids, item_ids])
        return predicted
```

```
def predict_single(model, user_id, item_id):
        predicted = model.predict([pd.Series([user_id]), pd.Series([item_id])])[0][0]
        return predicted
```

```
def evaluate(model, user_id_test, item_id_test, rating_test):
        test_preds = model.predict([user_id_test, item_id_test]).squeeze()
        mae = mean_absolute_error(test_preds, rating_test)
        print("Final test MAE: %0.3f" % mae)
        return {'mae': mae}
```

```
metrics = evaluate(model,user_id_test, item_id_test, rating_test)
```
