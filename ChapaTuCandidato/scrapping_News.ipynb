```
import math
import numpy as np
import pymongo
import re
import urllib.request as urllib2
import urllib.error
import datetime

from urllib.error import HTTPError
from time import sleep
from bs4 import BeautifulSoup

import urllib.parse


def getSoupObj(url):
    sleep(1.5)
    try:
        r = urllib2.urlopen(url)
    except UnicodeEncodeError as e:
        return None
    except HTTPError as err:
        if err.code == 404:
            return None
        else:
            raise
    html = r.read()
    htmltext= html.decode('utf-8')
    return BeautifulSoup(htmltext,'lxml')
```

```
def linksAndTexts(dfNews,dni,apePatMat):
  base_url="https://gestion.pe"
  base_urlQ = "https://gestion.pe/buscar/{}/todas/descendiente/?query={}"
  hdr = {'User-Agent': 'Mozilla/5.0'}

  url_news=base_urlQ.format(apePatMat,apePatMat)
  url_news = urllib.parse.urlsplit(url_news)
  url_news = list(url_news)
  url_news[2] = urllib.parse.quote(url_news[2])
  url_news[3] = urllib.parse.quote(url_news[3])
  url_news = urllib.parse.urlunsplit(url_news)
  req = urllib2.Request(url_news,headers=hdr)
  print(url_news)
  newsObj = getSoupObj(req)
  if newsObj==None:
    return
  titles= newsObj.findAll('a', {'class': 'story-item__title block overflow-hidden primary-font line-h-xs mt-10'})
  for title in titles[:5]:
    titleText=title.text
    newsLink=base_url+title['href']
    row=[dni,newsLink,titleText]
    dfNews.loc[len(dfNews)] = row
```

```
def scrapping_news(df):
  dfNews = pd.DataFrame(columns = ['dni', 'url_noticia','titulo_noticia'])
  for index,row in df.iterrows():
    print(index)
    apePatMat= row.ape_paterno + "+" + row.ape_materno
    dni=row.dni
    linksAndTexts(dfNews,dni,apePatMat)
  return dfNews
```

```
dfNews=scrapping_news(general_lima)
dfNews.to_csv('/content/drive/My Drive/ChapaTuCandidato/general_lima_news.csv') 
```
